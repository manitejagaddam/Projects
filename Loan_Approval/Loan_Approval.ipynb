```python
from scipy import stats

numerical_features = ['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 
                       'cibil_score', 'residential_assets_value', 'commercial_assets_value', 
                       'luxury_assets_value', 'bank_asset_value']


z_scores = stats.zscore(df[numerical_features])
```

```python
threshold = 3
# calculate the ouliers in the dataframe
outliers = np.any(np.abs(z_scores) > threshold, axis=1)

# Filter out the outliers
df_clean = df[~outliers]

# Verify the number of rows before and after outlier removal
print("Original number of rows:", len(df))
print("Number of rows after outlier removal:", len(df_clean))
```

```python
categorical_features = ['education', 'self_employed']
```

```python
cf = ColumnTransformer(transformers = [
    ('Encoding', OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ('Scaling', StandardScaler(), numerical_features)
], remainder = 'passthrough')
```

```python
X = df.drop(columns = ['loan_id', 'loan_status'], axis=1)
y = df['loan_status']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
decision_tree = DecisionTreeClassifier(max_depth = 7 )
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', cf),
    ('classifier', DecisionTreeClassifier(random_state=42, max_depth = 7))
])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 44)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
